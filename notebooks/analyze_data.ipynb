#UTK Face

The labels of each face image is embedded in the file name, formated like [age]_[gender]_[race]_[date&time].jpg

* [age] is an integer from 0 to 116, indicating the age
* [gender] is either 0 (male) or 1 (female)
* [race] is an integer from 0 to 4, denoting White, Black, Asian, Indian, and Others (like Hispanic, Latino, Middle Eastern).
* [date&time] is in the format of yyyymmddHHMMSSFFF, showing the date and time an image was collected to UTKFace

```python
import glob
import os
import pandas as pd


images = os.listdir('/content/UTKFaceDataset/UTKFace')
ages = []
genders = []
races = []

def parse_name_to_label(name):
  data = name.split('_')
  try:
    age = int(data[0])
    gender = int(data[1])
    race = int(data[2])
  except:
    print(name)
    age,gender,race = 0,0,0
  return age, gender, race


for img_name in images:
  age, gender, race = parse_name_to_label(img_name)
  ages.append(age)
  genders.append(gender)
  races.append(race)

data = {'images':images,
        'ages': ages,
        'genders': genders,
        'races': races}
df = pd.DataFrame(data)

df.to_csv('utk_annotations.csv', header=True, index=False)
df.shape
```

```python
import pandas as pd
import shutil
import cv2

from batch_face import RetinaFace
import numpy as np
import os
from glob import glob
from skimage import transform
from tqdm import tqdm
import random

random.seed(42)
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
```

```python
def alignt(imgname, mode = 1):
  detector = RetinaFace()
  img = cv2.imread(imgname)  
  faces = detector(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
  
  if len(faces) == 0:
    return img
  face = faces[0]
  box, landmark, score = face
  if mode == 1 :
    src = np.array(
        [
            [38.2946, 51.6963],
            [73.5318, 51.5014],
            [56.0252, 71.7366],
            [38.2946, 92.3655],
            [70.7299, 92.2041],
        ]
    )
  else:
    src = np.array(
        [
            [30.2946, 51.6963],
            [65.5318, 51.5014],
            [48.0252, 71.7366],
            [33.5493, 92.3655],
            [62.7299, 92.2041],
         ]
        )
  trans = transform.SimilarityTransform()
  res = trans.estimate(landmark, src)
  M = trans.params
  new_img = cv2.warpAffine(img, M[:2, :], dsize=(112, 112))
  return new_img
```

```python
dst_dir = '/content/faa_busmap_annotation'
os.makedirs(dst_dir, exist_ok=True)
store_dir = []
for i in range(0, 75, 5):
  dir_name = f"{str(i+1)}-{str(i+5)}"
  dir = f'{dst_dir}/{dir_name}'
  store_dir.append(dir_name)
  os.makedirs(dir, exist_ok=True)
```

```python
def age2store(age):
  if int(age) == 0:
    return store_dir[0]
  if int(age) < 0:
    raise Exception("age must be positive number")
  if int(age) > 70:
    return store_dir[-1]

  idx = age // 5
  rdd = 1 if age%5 == 0 else 0
  
  return store_dir[idx-rdd]
```

## UTK

```python
utk_df = pd.read_csv('/content/UTKFaceDataset/utk_annotations.csv')
utk_dir = '/content/UTKFaceDataset/UTKFace'
```

```python
ages = []
genders = []
image_path = []
classes = []
for i, (image, age, gender, race) in tqdm(utk_df.iterrows()):
  if int(race) == 2:
    dir_store = age2store(int(age))
    src=os.path.join(utk_dir, image)
    dst=f"{dst_dir}/{dir_store}/{image}"
    shutil.copy(src=src,dst=dst)
    image_path.append(f"{dir_store}/{image}")
    ages.append(int(age))
    genders.append(int(gender))
    cls = store_dir.index(dir_store)
    classes.append(cls)

data = {'image_name': image_path,
        'age': ages,
        'gender': gender,
        'classes': classes}
df = pd.DataFrame(data)
df.to_csv('annotation_faa_busmap_utk.csv', index=False)
```

## FairFace

```python
train_df = pd.read_csv('/content/drive/MyDrive/FaceAgeGender/Dataset/FairFace/fairface_label_train.csv')
val_df = pd.read_csv('/content/drive/MyDrive/FaceAgeGender/Dataset/FairFace/fairface_label_val.csv')
fairface_dir = '/content/fairface_margin025'

fairface_df = pd.concat([train_df, val_df])
fairface_df.shape
```

```python
fairface_df['age'].unique()
```

```python
def fairface_age2storedir(age):
  if '70' in age:
    return store_dir[-1], 71
  
  start, end = age.split('-')
  start, end = int(start), int(end)
  age = random.randint(start,end)
  return age2store(age), age
```

```python
for i, (file, age, gender, race, service_test) in tqdm(fairface_df.iterrows()):
  final_gender = 0 if gender=='Male' else 1
  fairface_df.at[i,'gender'] = final_gender
```

```python
ages = []
genders = []
image_path = []
classes = []

for i, (image, age, gender, race, service_test) in tqdm(fairface_df.iterrows()):
  if 'Asian' in race:
    dir_store, fn_age = fairface_age2storedir(age)
    src = os.path.join(fairface_dir, image)
    image_name = image.split('/')[-1]
    dst = f"{dst_dir}/{dir_store}/{image_name}" 
    shutil.copy(src=src, dst=dst)
    image_path.append(f"{dir_store}/{image_name}")
    ages.append(int(fn_age))
    genders.append(int(gender))
    cls = store_dir.index(dir_store)
    classes.append(int(cls))
    

data = {'image_name': image_path,
        'age': ages,
        'gender': gender,
        'classes': classes}
df = pd.DataFrame(data)
df.to_csv('annotation_faa_busmap_fairface.csv', index=False)
```

```python
def add_class_index(input_csv):
  df = pd.read_csv(input_csv)
  classes = []
  print("process: ", input_csv)
  for i, (image_name, age, gender) in tqdm(df.iterrows()):
    cls = image_name.split('/')[0]
    cls = store_dir.index(cls)
    classes.append(cls)
  df['classes'] = classes
  df.to_csv(f"{input_csv}", index=False)
```

```python
def convert_value_to_int(input_csv):
  df = pd.read_csv(input_csv)
  print("process: ", input_csv)
  for i, (image_name, age, gender, classes) in tqdm(df.iterrows()):
    new_age = int(age)
    new_gender = int(gender)
    new_classes = int(classes)
    df.at[i,'age'] = new_age
    df.at[i,'gender'] = new_gender
    df.at[i,'classes'] = new_classes
  
  df.to_csv(f'{input_csv}', index=False)
```
